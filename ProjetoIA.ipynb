```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fatores_estresse.csv")
```

```python
x = df.drop('How would you rate you academic performance üë©‚Äçüéì?', axis=1)
y = df['How would you rate you academic performance üë©‚Äçüéì?']
```

```python
scaler = MinMaxScaler()
df["performance_academica_norm"] = scaler.fit_transform(df[["How would you rate you academic performance üë©‚Äçüéì?"]])
```

```python
X_scaled = scaler.fit_transform(x)
```

```python
print(df[["How would you rate you academic performance üë©‚Äçüéì?", "performance_academica_norm"]])
```

```python
print("Antes da transforma√ß√£o:")
```

```python
print(df)
```

```python
df_estresse_remover = df[df["How would you rate your stress levels?"] != 5]
```

```python
print("Depois de remover linhas com o m√°ximo de n√≠vel de estresse:")
```

```python
print(df_estresse_remover['How would you rate your stress levels?'])
```

```python
X_train, X_temp, Y_train, Y_temp = train_test_split(
X_scaled, y, test_size=0.30, random_state=42
)
```

```python
X_val, X_test, Y_val, Y_test = train_test_split(
X_temp, Y_temp, test_size=0.50, random_state=42
)
```

```python
print("Quantidade de dados de treino:", len(X_train))
print("Quantidade de dados de valida√ß√£o", len(X_val))
print("Quantidade de dados de teste", len(X_test))
```

```python
model = RandomForestClassifier(n_estimators=200, random_state=42)
```

```python
model.fit(X_train, Y_train)
```

```python
Y_val_pred = model.predict(X_val)
```

```python
val_accuracy = accuracy_score(Y_val, Y_val_pred)
```

```python
print("Precis√£o da valida√ß√£o", val_accuracy)
```

```python
Y_test_pred = model.predict(X_test)
```

```python
test_accuracy = accuracy_score(Y_test, Y_test_pred)
```

```python
print("Precis√£o do teste:", test_accuracy)
```

```python
cm = confusion_matrix(Y_test, Y_test_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
```

```python
disp.plot(cmap='Blues')
```
